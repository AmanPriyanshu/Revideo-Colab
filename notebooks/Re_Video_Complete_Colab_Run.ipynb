```python
import subprocess
import time

# Define the project name and path
project_name = "project"
project_path = "project"
npm_init_command = ["npm", "init", "@revideo@latest"]
process = subprocess.Popen(npm_init_command, stdin=subprocess.PIPE, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True)
time.sleep(1)
process.stdin.write(project_name + "\n")
process.stdin.flush()
time.sleep(1)
process.stdin.write(project_path + "\n")
process.stdin.flush()
process.stdin.close()
return_code = process.wait()
print(f"Return code: {return_code}")
```
